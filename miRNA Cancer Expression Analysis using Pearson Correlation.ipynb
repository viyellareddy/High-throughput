```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import pearsonr

# Load Matrix 1
df1 = pd.read_csv('matrix1.txt', sep='\t')
data1 = pd.DataFrame(df1)

# Set index as miRNA and drop unnecessary column if it exists
data1.set_index('miRNA', inplace=True)
if 'Unnamed: 13' in data1.columns:
    data1.drop('Unnamed: 13', axis=1, inplace=True)

# Compute Pearson correlation matrix for Matrix 1
matrix1 = data1.corr()
print(matrix1)

# Visualize correlation matrix as heatmap
sns.heatmap(matrix1)

```

```python
# Load Matrix 2
df2 = pd.read_csv('matrix2.txt', sep='\t')
data2 = pd.DataFrame(df2)

# Set index as miRNA and drop unnecessary column if it exists
data2.set_index('miRNA', inplace=True)
if 'Unnamed: 13' in data2.columns:
    data2.drop('Unnamed: 13', axis=1, inplace=True)

# Compute Pearson correlation matrix for Matrix 2
matrix2 = data2.corr()
print(matrix2)

# Visualize correlation matrix as heatmap
sns.heatmap(matrix2)
```

```python
# Compare the two correlation matrices by flattening them
matrix1_flat = matrix1.values.flatten()
matrix2_flat = matrix2.values.flatten()

# Compute Pearson correlation between the two matrices
correlation_between_matrices, _ = pearsonr(matrix1_flat, matrix2_flat)
print("Correlation between Matrix 1 and Matrix 2:", correlation_between_matrices)
```
